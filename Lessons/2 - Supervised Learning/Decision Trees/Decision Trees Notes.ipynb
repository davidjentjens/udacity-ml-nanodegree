# Decision Trees Notes

#### Entropy formula for 2 different elements:

```python
import math 

def entropy(m, n):
    entropy_m = -(m/(m+n)) * math.log(m/(m+n),2)
    entropy_n = -(n/(m+n)) * math.log(n/(m+n),2)
    return entropy_m + entropy_n
                
entropy(4,10)
```

#### Entropy formula for n elements:

```python
import math 

def entropy(balls):
    total_balls = sum(balls)
    total_entropy = []
    for i in range(len(balls)):
        ball_probability = balls[i]/total_balls
        ball_entropy = -ball_probability * math.log(ball_probability,2)
        total_entropy.append(ball_entropy)
    return sum(total_entropy)
                
entropy([8,3,2])
```

## Bugs dataset problem

Which of the following splitting criteria provides the most information gain for discriminating Mobugs from Lobugs?

- Color = Brown
- Color = Blue
- Color = Green
- Length < 17.0mm
- Length < 20.0mm

```python
import pandas as pd
```

```python
df = pd.read_csv('ml-bugs.csv')
df
```

#### First test: **Color = Brown**

```python
division_a = df.loc[df['Color'] == 'Brown']
division_b = df.loc[df['Color'] != 'Brown']

mobug_ocurrances = (df.Species == 'Mobug').sum()
lobug_ocurrances = (df.Species == 'Lobug').sum()
entropy_before = entropy([mobug_ocurrances, lobug_ocurrances])

mobug_ocurrances_a = (division_a.Species == 'Mobug').sum()
lobug_ocurrances_a = (division_a.Species == 'Lobug').sum()
print('Division A')
print('\n', division_a, '\n')
print('Mobug ocurrances:', mobug_ocurrances_a)
print('Lobug ocurrances:', lobug_ocurrances_a, '\n')
entropy_division_a = entropy([mobug_ocurrances_a, lobug_ocurrances_a])

mobug_ocurrances_b = (division_b.Species == 'Mobug').sum()
lobug_ocurrances_b = (division_b.Species == 'Lobug').sum()
print('Division B')
print('\n', division_b, '\n')
print('Mobug ocurrances:', mobug_ocurrances_b)
print('Lobug ocurrances:', lobug_ocurrances_b, '\n')
entropy_division_b = entropy([mobug_ocurrances_b, lobug_ocurrances_b])

total_ocurrances_a = mobug_ocurrances_a + lobug_ocurrances_a
total_ocurrances_b = mobug_ocurrances_b + lobug_ocurrances_b
total_ocurrances = total_ocurrances_a + total_ocurrances_b

entropy_division_mean = (total_ocurrances_a * entropy_division_a + total_ocurrances_b * entropy_division_b) / total_ocurrances

information_gain = entropy_before - entropy_division_mean

print('Entropy before:', entropy_before, '\n')
print('Entropy division A:', entropy_division_a)
print('Entropy division B:', entropy_division_b)
print('Entropy after:', entropy_division_mean, '\n')
print('Information gain:', information_gain)
```
