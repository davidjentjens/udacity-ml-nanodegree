```python
import math

def calc_weight(accuracy):
    return math.log(accuracy / (1 - accuracy))
```

```python
accuracy_1 = 7/8
accuracy_2 = 4/8
accuracy_3 = 2/8

print('Weight of model 1:', calc_weight(accuracy_1))
print('Weight of model 2:', calc_weight(accuracy_2))
print('Weight of model 3:', calc_weight(accuracy_3))
```
